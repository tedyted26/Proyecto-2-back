```python
from textblob import TextBlob
polaridadTotal = 0
positivos = 0
negativos = 0
neutros = 0
for tweet in listatextos:    
    try:
        texto = TextBlob(tweet)
        traduccion = TextBlob(str(texto.translate(to='en')))
        print(texto)
        #print(traduccion)
        if (traduccion.sentiment[0]>0):
            positivos += 1
            polaridadTotal = polaridadTotal+traduccion.sentiment[0]
        elif (traduccion.sentiment[0]<0):
            negativos += 1
            polaridadTotal = polaridadTotal+traduccion.sentiment[0]
        elif (traduccion.sentiment[0]==0):
            neutros += 1
        print(traduccion.sentiment[0])
    except:
        print("An exception occurred")
        
#print(polaridadTotal)
#print("Numero a hacer media "+str(positivos+negativos))
print("Polaridad media: "+str(polaridadTotal/(positivos+negativos)))
print("Negativos: " + str(negativos))
print("Positivos: " + str(positivos))
print("Neutros: " + str(neutros))
```
