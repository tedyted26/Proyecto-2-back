```python
listaTweets = tweepy.Cursor(api.search_tweets, q='barrio del pilar').items(50)
listatextos = []
for tweet in listaTweets:
    print(tweet.text)
    listatextos.append(tweet.text)
```

```python
from textblob import TextBlob
polaridadTotal = 0
positivos = 0
negativos = 0
neutros = 0
for tweet in listatextos:    
    texto = TextBlob(tweet)
    traduccion = TextBlob(str(texto.translate(to='en')))
    print(texto)
    print(traduccion)
    if (traduccion.sentiment[0]>0):
        positivos += 1
    elif (traduccion.sentiment[0]<0):
        negativos += 1
    elif (traduccion.sentiment[0]==0):
        neutros += 1
        
    print(traduccion.sentiment[0])
    polaridadTotal = polaridadTotal+traduccion.sentiment[0]
    
print("Polaridad media: "+str(polaridadTotal/len(listatextos)))
print("Negativos: " + str(negativos))
print("Positivos: " + str(positivos))
print("Neutros: " + str(neutros))
```
